```python
import pandas as pd
import numpy as np

# Example data
data = {'Math': [90, 80, 70, 60, 50]}
df = pd.DataFrame(data)

# Converting the 'Math' column to a NumPy array
math_scores = df['Math'].to_numpy()

# Calculating standard deviation using pandas (sample standard deviation)
std_pandas = df['Math'].std()

# Calculating standard deviation using NumPy (population standard deviation)
std_numpy = np.std(math_scores)

print("Standard Deviation with pandas:", std_pandas)
print("Standard Deviation with NumPy:", std_numpy)
```

```python
std_numpy_sample = np.std(math_scores, ddof=1)  # ddof=1 makes it a sample standard deviation
std_pandas_population = df['Math'].std(ddof=0)  # ddof=0 makes it a population standard deviation
print("Standard Deviation with pandas population:", std_pandas_population)
print("Standard Deviation with NumPy:", std_numpy)
```

## difference
**the difference in standard deviation results is primarily due to the default methods of calculationNumPy uses the population standard deviation by default, while pandas uses the sample standard deviation. Adjusting the ddof (Delta Degrees of Freedom) parameter can align the calculations to produce the same results**

```python
# Convert DataFrame column to NumPy array
math_scores = df['Math'].to_numpy()

# Mean
mean_math = np.mean(math_scores)
print(f"Mean of Math scores: {mean_math}")

# Median
median_math = np.median(math_scores)
print(f"Median of Math scores: {median_math}")

# Mode using SciPy
mode_math = sp.stats.mode(math_scores)[0]
print(f"Mode of Math scores: {mode_math}")

# Standard Deviation
std_math = np.std(math_scores)
print(f"Standard Deviation of Math scores: {std_math}")
```

## KDE


KDE stands for Kernel Density Estimation. It's a non-parametric way to estimate the probability density function of a random variable. In simpler terms, KDE smooths out the data to show the distribution more clearly, especially when you have a limited number of observations.
KDE is a useful tool for visualizing the distribution of data, offering a smooth curve that complements the discrete bins of a histogram. By adding kde=True in your sns.histplot function, you enhance the histogram with a KDE line, giving a clearer picture of the data's underlying distribution.

# 2.5 Skewness

## 2.5.1 Explain
Skewness: A measure of the asymmetry of the probability distribution. Positive skew indicates a distribution with a longer tail on the right; negative skew indicates a longer tail on the left.

FM 4: Write Skewness Formula and also how to analysis skew??

Interpret Skewness Value:

Positive Skewness: If skewness () is greater than zero, it indicates that the distribution has a tail on the right side (right-skewed). This means there are more data points on the left side of the distribution with a few extreme values on the right side.

Negative Skewness: If skewness () is less than zero, it indicates that the distribution has a tail on the left side (left-skewed). This means there are more data points on the right side of the distribution with a few extreme values on the left side.

Zero Skewness: A skewness value of zero indicates that the distribution is perfectly symmetrical around the mean. This means the left and right tails of the distribution are equal in length, and the data is evenly distributed around the mean.

Skewness Formula
Skewness () is a measure of asymmetry in a distribution. It is typically calculated using the third standardized moment (3) normalized by the cube of the standard deviation ():

$1=\frac{3}{3}$



Where:

3 is the third central moment (moment about the mean).

  is the standard deviation

## 2.5.2 Code & Visualization

```python
for column in df.select_dtypes(include=[np.number]).columns:
    print(f"Analyzing column: {column}")

    # Descriptive statistics
    mean_col = df[column].mean()
    median_col = df[column].median()
    std_col = df[column].std()
    skewness_col = df[column].skew()
    print(f"Mean: {mean_col}")
    print(f"Median: {median_col}")
    print(f"Standard Deviation: {std_col}")
    print(f"Skewness: {skewness_col}")

    # Histogram
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    sns.histplot(df[column], kde=True)
    plt.title(f"Histogram of {column}")

    # Box plot
    plt.subplot(1, 2, 2)
    sns.boxplot(y=df[column])
    plt.title(f"Box Plot of {column}")
    plt.show()
```
